# RÃ©seaux de neurones convolutifs

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from scipy.signal import convolve2d
from sklearn.datasets import load_sample_image
```

```python
def plot_gray_image(img):
    "Affiche une image en niveau de gris"
    plt.imshow(img, cmap="gray")
    plt.axis("off")
```

## Filtres communs pour le traitement d'images

### Filtre moyen

```python
flower_image = load_sample_image("flower.jpg")

plt.imshow(flower_image)
plt.show()
```

```python
import cv2
```

```python
gray = cv2.cvtColor(flower_image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
plt.axis("off")
plt.show()
```

```python
mean = 0
var = 100
sigma = var**0.5
row, col = gray.shape

# CrÃ©ation du bruit gaussien
gauss = np.random.normal(mean, sigma, (row, col))

gray_noisy = gray + gauss

plt.figure(figsize=(20, 30))
plt.subplot(121)
plot_gray_image(gray)

plt.subplot(122)
plot_gray_image(gray_noisy)

plt.show()
```

```python
# Filtre moyen

mean_filter = np.array([[1,1,1], [1,1,1], [1,1,1]]) / float(9)
mean_filter
```

```python
new_image = convolve2d(gray, mean_filter, mode="same")

plt.figure(figsize=(20, 30))
plt.subplot(121)
plot_gray_image(gray_noisy)

plt.subplot(122)
plot_gray_image(new_image)

plt.show()
```

```python
plt.figure(figsize=(20, 30))
plt.subplot(121)
plot_gray_image(gray)

plt.subplot(122)
plot_gray_image(new_image)

plt.show()
```
